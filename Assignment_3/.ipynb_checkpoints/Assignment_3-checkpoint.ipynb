Perform a univariate analysis on all the categorical data of the penguins dataset. Commit the notebook to your portfolio when you're finished.

```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins['species'].value_counts(dropna=False).plot(kind='pie')
```

```python
penguins['island'].value_counts(dropna=False).plot(kind='pie')
```

```python
penguins['sex'].value_counts(dropna=False).plot(kind='bar')
```

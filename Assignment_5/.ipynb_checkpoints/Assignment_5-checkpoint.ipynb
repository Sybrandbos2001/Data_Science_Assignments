```python
import pandas as pd
```

```python
lifeExpectancy = pd.read_csv('lifeExpectancyAtBirth.csv', sep=',')
```

```python
lifeExpectancy.head()
```

```python
FilteredList = lifeExpectancy[(lifeExpectancy['Period'] == 2019) & (lifeExpectancy['Dim1'] == "Both sexes")]
```

```python
FilteredList.head()
```

```python
FilteredList[(FilteredList['First Tooltip'] >= 84)]
```

### Je kunt hieruit opmaken dat enkel in Japan de gemiddelde levensverwachting meer dan 84 jaar is.

```python
FilteredList.sort_values('First Tooltip', ascending = False)
```

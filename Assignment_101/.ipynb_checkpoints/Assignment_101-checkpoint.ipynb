### Verzamelen van gegevens: Als data scientist zul je eerst gegevens moeten verzamelen voordat je onderzoeken kunt doen en analyses. Dit kan van alles zijn, denk aan namen van studenten (nominale data) of het aantal inwoners van een bepaalde stad (discrete data). Deze data verzamel je en leg je vast in een formaat dat rijen en kolommen heeft. In dit geval is er gekozen voor een Pokemon dataset.

### Onderzoeken: In dit geval was de vraag "Hoeveel pokemons zijn er van elk type?". Op basis van deze vraag ga je onderzoek doen. Je kijkt welke kolommen hier van toepassing zijn en denkt na hoe je dit gaat weergeven. In dit geval een staafdiagram met op de X-as het type en de Y-as de hoeveelheid.

### Analyseren: Als datascientist krijg je soms te maken met gigantische datasets. Om deze data stap voor stap door te gaan ben je heel lang bezig. Daarom gebruikt een data scientist vaak clusters om snel te kunnen analyseren of er mogelijke verbanden zitten tussen data. In afbeelding hierboven word er gekeken of penguins bepaalde uiterlijke kenmerken hebben die mogelijk overeenkomen per soort.

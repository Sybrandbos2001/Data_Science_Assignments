### Verzamelen van gegevens: In assignment 4 ben ik opzoek gegaan naar een dataset om te gaan gebruiken in mijn portfolio. Ik heb hierbij de dataset geanalyseerd op de datatypen die kolommen hebben. Zo moest de dataset 2 kolommen met catogorische data en 2 kolommen met numerieke data.

### Analyseren: Kort voordat je aan de slag gaat met een dataset doe je een korte snelle analyse of de dataset die je wilt gebruiken wel overeenkomt met degene die jij in gedachten hebt.   

### Vaststellen van vraag of informatiebehoefte: Als je een onderzoek uitvoert heb je een vraagstelling of informatiebehoefte. Wat wil je nu precies weten, als data scientist is het belangrijk dat je informatiebehoefte of vraagstelling zo concreet mogelijk is. Welke kolommen ga ik gebruiken, wat zijn mijn verwachtingen?

### Onderzoek doen: Als je informatiebehoefte helder is kun je onderzoek doen. Je haalt data op, transformeert deze waar nodig, zoekt correlaties en laat dit alles zien in plotten of tabellen.

### Concluderen: Als data scientist moet je conclusies kunnen trekken uit jouw gedane onderzoek. Je moet kunnen stellen of er is voldaan aan je informatie behoefte of vraagstelling. 

### Machinelearning : Als datascientist moet je ook methodes kunnen bouwen die 'leren'. Zo moet je een training set kunnen bepalen die de uitkomst van een machine niet te veel beinvloed. Als je bijvoorbeeld de gemiddelde lengte van een inwoner van Nederland wilt berekenen op gewicht moet je ervoor zorgen dat er evenveel mannen als vrouwen worden meegenomen in de testset. Hier moet je vooraf over nadenken. 

### Ook moet je van te voren bepalen op welke eisen de computer moet gaan beoordelen, zogenoemde features.

### Beoordelen van onderzoek : Als datascientist moet je ook kritisch kijken naar je onderzoek, hoe accuraat kun je dit nou noemen. Als je een dataset hebt met 2 rijen zal dit onderzoek een stuk minder accuraat zijn dan met een dataset van 10 duizend rijen. Als data scientist kun je met verschillende wiskundige formules berekenen hoe accuraat jouw onderzoek nou precies is geweest.

```python
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.groupby('species').mean()
```

```python
penguins.groupby('island').mean()
```

```python
penguins.groupby('sex').mean()
```

```python
sns.barplot(y="bill_length_mm", x="species", data=penguins, ci = 99.99)
```

```python
sns.barplot(y="bill_depth_mm", x="species", data=penguins, ci = 99.99)
```

```python
sns.barplot(y="body_mass_g", x="species", data=penguins, ci = 99.99)
```

```python
sns.barplot(y="flipper_length_mm", x="species", data=penguins, ci = 99.99)
```
